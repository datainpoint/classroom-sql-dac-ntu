## 什麼是 SQL？

## SQL（發音為 ess-que-ell 或 sequel）全名為 Structured Query Language，是一個能夠針對關聯式資料庫管理系統進行「資料操作」的語言。

## SQL 按照使用目的可以再細分為三類

1. **資料操作語言**（Data Manipulation Language, DML）
2. 資料定義語言（Data Definition Language, DDL）
3. 資料控制語言（Data Control Language, DCL）

## 解釋定義中出現的名詞

- 什麼是**資料操作**？
- 什麼是資料庫？
- 什麼是資料庫管理系統？

## 「資料操作」涵蓋了 CRUD 這四個動詞：

- 創造 **C**reate
- 查詢 **R**ead
- 更新 **U**pdate
- 刪除 **D**elete

## 舉例來說，在使用任何的網頁或手機應用程式時，我們的滑鼠點擊與手勢觸控都會被轉換成 CRUD：

- 創造 **C**reate：發佈新的動態。
- 查詢 **R**ead：瀏覽追蹤對象的動態。
- 更新 **U**pdate：編輯先前發佈動態的內容。
- 刪除 **D**elete：撤掉先前所發佈的動態。

## DBMS 全名為 DataBase Management System，透過資料庫管理系統，SQL 將能「自動化」且「規模化」地對關聯式資料庫進行 CRUD 的「資料操作」。

## 自動化的體現

不需要透過人工、客服就能夠發佈新的動態、瀏覽追蹤對象的動態、編輯先前發佈動態的內容與撤掉先前所發佈的動態。

## 規模化的體現

能夠讓成千上萬個使用者同時發佈新的動態、瀏覽追蹤對象的動態、編輯先前發佈動態的內容與撤掉先前所發佈的動態。

## 資料庫管理系統可以分為兩大類：

1. 關聯式資料庫管理系統（RDBMS, Relational Database Management System）
2. 非關聯式資料庫管理系統（NoSQL DBMS, Not only SQL Database Management System）

## RDBMS vs. NoSQL DBMS

- 結構化 vs. 非結構化。
- 表格 vs. 文件、鍵值。
- 效率 vs. 彈性。

## `SELECT`、`FROM` 與 `LIMIT`

- `SELECT` 是「選擇」欄位。
- `FROM` 是指定「從」哪個資料表查詢資料。
- `LIMIT` 是讓資料查詢結果顯示指定的前幾列。
- `*` 表示「所有」。
- `;` 表示一段 SQL 的結束。

## 儲存在資料庫中的資料表是有兩個維度的資料結構

- 第一個維度稱為觀測值（Observations），有時亦稱為列（Rows）。
- 第二個維度稱為變數（Variables），有時亦稱為欄（Columns）。

## 基礎 SQL 語法：衍生計算欄位

## 在沒有 `FROM` 的情況下，`SELECT` 也能夠單獨存在於一段 SQL 之中

````mysql
SELECT 'Hello, SQL!' AS hello_sql; -- 單獨存在的 SELECT
````

## 在 `SELECT` 後面除了指定資料表的欄位，也能夠擺放常數，常見的常數類型有：

- 整數（INTEGER）。
- 帶小數點的浮點數（REAL）。
- 用單引號或雙引號標記的文字（TEXT）。
- 空值、遺漏值（NULL）。

## 使用單獨存在的 `SELECT` 將常見的常數類型顯示出來

````mysql
SELECT 5566 AS an_integer,
       5566.0 AS a_real,
       'Hello, SQL!' AS a_text,
       NULL AS a_null;
````

## 使用函數 `TYPEOF()` 將常數類型顯示出來

````mysql
SELECT TYPEOF(5566) AS typeof_an_integer,
       TYPEOF(5566.0) AS typeof_a_real,
       TYPEOF('Hello, SQL!') AS typeof_a_text,
       TYPEOF(NULL) AS typeof_a_null;
````

## 這表示即便在資料表中沒有儲存常數，依然能在 SQL 中加入這些常數

````mysql
SELECT runtime,
       60.0 AS minutes_per_hour
  FROM movies
 LIMIT 5;
````

## 針對整數（INTEGER）與帶小數點的浮點數（REAL）可以使用數值運算符衍生計算欄位

|數值運算符|作用描述|
|--------|-------|
|`+`|相加|
|`-`|相減|
|`*`|相乘|
|`/`|相除|
|`%`|回傳餘數|
|`()`|優先運算|

## 運用數值運算符

````mysql
SELECT runtime AS runtime_in_minutes,
       runtime / 60 AS runtime_in_hours,
       runtime % 60 AS runtime_in_mins
  FROM movies
 LIMIT 5;
````

## 數值運算時要特別注意優先順序，例如先乘除後加減

\begin{equation}
Celsius_{(^{\circ} C)} = (Fahrenheit_{(^{\circ} F)} - 32) \times \frac{5}{9}
\end{equation}

````mysql
SELECT 212 - 32 * 5/9 AS wrong_conversion,      -- 錯誤的華氏轉攝氏
       (212 - 32) * 5/9 AS correct_conversion;  -- 正確的華氏轉攝氏
````

## 在使用 `/` 相除兩個整數的時候要特別注意所衍生的欄位依然會以整數類型存在

````mysql
SELECT 1 / 2 AS an_integer;
````

## 若希望衍生欄位是浮點數類型，要為分子或者分母加上小數點

````mysql
SELECT 1.0 / 2 AS a_real;
````

## 針對文字（TEXT）可以使用文字運算符衍生計算欄位

`||`：兩個垂直線（可透過 `Shift + \` 按出來）能夠連接文字。 

````mysql
SELECT title || ', directed by ' || director AS title_director
  FROM movies
 LIMIT 5;
````

## 基礎 SQL 語法：以 `ORDER BY` 排序查詢結果

## 加入 `ORDER BY` 能夠以指定變數的大小來排序查詢結果

```sql
SELECT column_names
  FROM table_name
 ORDER BY column_names;
```

## 排序方式有兩種：

1. 遞增（或稱升冪）排序。
2. 遞減（或稱降冪）排序。

## 預設的排序方式為遞增

````mysql
SELECT *
  FROM movies
 ORDER BY runtime;
````

## 如果想「明確」表達遞增排序，可以加上保留字 `ASC`（Ascending），但這麼做會顯得有些多餘

````mysql
SELECT *
  FROM movies
 ORDER BY rating ASC;
````

## 如果指定排序的變數是文字（TEXT）類型，遞增的規範是由 A 到 Z

````mysql
SELECT *
  FROM movies
 ORDER BY director;
````

## 如果希望讓查詢結果遞減排序，必須加上保留字 `DESC`（Descending）

````mysql
SELECT *
  FROM movies
 ORDER BY runtime DESC;
````

## 在 `ORDER BY` 後，可以指定多個變數搭配各自的排序方式

指定多個變數排序時，會依照「先指定」的變數排序，在「先指定」的變數排序相同的情況下再依據「後指定」的變數排序。

````mysql
SELECT *
  FROM movies
 ORDER BY release_year DESC, -- 先指定 release_year 遞減排序
          rating;            -- 後指定 rating 遞增排序
````

## 結合 `ORDER BY` 與 `LIMIT`，就可以進行「前 `m` 高」或「前 `m` 低」的資料分析

```sql
-- 「前 m 低」的資料分析
SELECT column_names
  FROM table_name
 ORDER BY column_names
 LIMIT m;
```

```sql
-- 「前 m 高」的資料分析
SELECT column_names
  FROM table_name
 ORDER BY column_names DESC
 LIMIT m;
```

````mysql
SELECT *
  FROM movies
 ORDER BY release_year
 LIMIT 10;
````

````mysql
SELECT *
  FROM movies
 ORDER BY release_year DESC
 LIMIT 11;
````

## 撰寫條件之前，我們要暸解兩個觀念：

1. 關係運算符：能夠產生布林（Boolean）的運算符號。
2. 布林（Boolean）：用來表示比較結果的兩個值（假、真）。

## 基礎關係運算符

|關係運算符|作用描述|
|--------|-------|
|`=`|相等|
|`!=`|不相等|
|`>`|大於|
|`<`|小於|
|`>=`|大於等於|
|`<=`|小於等於|

## 比較結果為「假」的布林，SQLite 以 `0` 表示

````mysql
SELECT 5566 != 5566 AS bool_false;
````

## 比較結果為「真」的布林，SQLite 以 `1` 表示

````mysql
SELECT 5566 = 5566 AS bool_true;
````

## 針對資料表中的欄位使用關係運算符會在每列觀測值都對應生成一個布林

````mysql
SELECT release_year = 1994 AS bool_values
  FROM movies
 LIMIT 10;
````

## 若是在 `WHERE` 後使用關係運算符撰寫條件，會篩選出布林為真（`1`）的觀測值

````mysql
SELECT release_year = 1994 AS bool_values
  FROM movies
 WHERE release_year = 1994;
````

````mysql
SELECT *
  FROM movies
 WHERE release_year = 1994;  -- 篩選 1994 年上映的電影
````

## 關係運算符也可以應用在文字類型的變數

````mysql
SELECT *
  FROM movies
 WHERE director = 'Christopher Nolan';
````

## 特徵比對

## 除了基礎關係運算符，對文字類型的變數撰寫條件時，還能夠使用具備特徵比對（Pattern matching）性質的關係運算符 `LIKE`

## 使用 `LIKE` 關係運算符的時候需要搭配萬用字元（Wildcards）

|萬用字元|作用描述|
|-------|------|
|`%`|表示任意文字，包含空字串|
|`_`|表示剛好一個文字|

````mysql
SELECT *
  FROM movies
 WHERE title LIKE '%Dark Knight%';
````

````mysql
SELECT *
  FROM movies
 WHERE title LIKE '____Dark_______';
````

## 當 `WHERE` 後的條件有多個的時候，必須使用邏輯運算符結合這些條件

## 基礎的邏輯運算符有：

- `AND` 結合條件的交集。
- `BETWEEN` 結合數值比較條件的交集。
- `OR` 結合條件的聯集。
- `IN` 結合條件的聯集。
- `NOT` 反轉真假。

## 使用 `AND` 結合兩個條件時，要兩條件皆為真才會判斷為真，其餘狀況均為假

````mysql
SELECT rating >= 8.8 AS condition_1,
       rating <= 9.0 AS condition_2,
       rating >= 8.8 AND rating <= 9.0 AS condtion_1_and_condition_2
  FROM movies
 LIMIT 15;
````

````mysql
SELECT *
  FROM movies
 WHERE rating >= 8.8 AND
       rating <= 9.0;  -- 評等介於 8.8 與 9.0 之間的電影
````

## 以 `AND` 結合數值比較條件時，更推薦使用 `BETWEEN`

````mysql
SELECT *
  FROM movies
 WHERE rating BETWEEN 8.8 AND 9.0;  -- 評等介於 8.8 與 9.0 之間的電影
````

## 使用 `OR` 結合兩個條件時，要兩者皆為假才為假，其餘狀況均為真

````mysql
SELECT director = 'Christopher Nolan' AS condition_1,
       director = 'Steven Spielberg' AS condition_2,
       director = 'Christopher Nolan' OR director = 'Steven Spielberg' AS condition_1_or_condition_2
  FROM movies
 LIMIT 10;
````

````mysql
SELECT *
  FROM movies
 WHERE director = 'Christopher Nolan' OR
       director = 'Steven Spielberg';
````

## 以 `OR` 結合比較條件時，更推薦使用 `IN`

````mysql
SELECT *
  FROM movies
 WHERE director IN ('Christopher Nolan', 'Steven Spielberg');
````

## 使用 `NOT` 將條件的比較結果反轉，亦即真假互換

````mysql
SELECT director = 'Christopher Nolan' AS condition_1,
       director = 'Steven Spielberg' AS condition_2,
       NOT (director = 'Christopher Nolan' OR director = 'Steven Spielberg') AS condition_1_or_condition_2
  FROM movies
 LIMIT 10;
````

````mysql
SELECT *
  FROM movies
 WHERE director NOT IN ('Christopher Nolan', 'Steven Spielberg');
````

## 要判斷是否為遺漏值，必須使用 `IS NULL` 作為關係運算符
